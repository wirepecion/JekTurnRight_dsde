```python
station_df = pd.read_csv('data/raw/station_metadata.csv')
water_df = pd.read_csv('data/external/rain_water_merged_2022_to_2024.csv')
```

```python

# clean Station_df
station_df['DistrictName'] = station_df['DistrictName'].replace({
    'ป้อมปราบฯ': 'ป้อมปราบศัตรูพ่าย',
    'ราษฏร์บูรณะ': 'ราษฎร์บูรณะ',
})
station_df = station_df[station_df['DistrictName'] != 'อำเภอเมืองสมุทรปราการ']
if len(station_df['StationCode'][0]) == 9: 
    station_df['StationCode'] = station_df['StationCode'].str[3:]
station_df = station_df[station_df['StationCode'] != 'PYT.02']

display(df)
display(station_df)
display(water_df)
```

```python
#clean water
water_df['Date'] = pd.to_datetime(water_df['Date'],format='mixed')
imp = SimpleImputer(strategy='most_frequent')
# water_df = water_df.drop(columns='NKM.03',axis=1)
station_list = ['KSW.02','HKG.02','MBR.03','LSI.02','WTL.02']
# print(water_df[station_list].dtypes)
# print(water_df[ station_list]) 

# print(water_df.isna().sum().sort_values().tail(20))

water_df[station_list] = imp.fit_transform(water_df[station_list])

water_long = water_df.melt(id_vars='Date', var_name='StationCode', value_name='water')
display(water_long.isna().sum())
```

```python
from sklearn.neighbors import BallTree

def make_station_info_nearest(df: pd.DataFrame, station_df: pd.DataFrame):

    df = df.copy()
    n_rows = len(df)

    station_map = (
        station_df.groupby('DistrictName')
        .apply(lambda g: g[['StationCode', 'Latitude', 'Longitude']].to_dict('records'))
        .to_dict()
    )

    df['StationList'] = df['district'].map(station_map)
    df['StationCode'] = np.nan

    df_rad = np.radians(df[['latitude','longitude']].to_numpy())
    positions = np.arange(n_rows)

    for district, stations in station_map.items():
        # Positional indices of rows in this district
        idx = positions[df['district'].to_numpy() == district]
        if len(idx) == 0:
            continue
        
        if len(stations) == 1:
            df.loc[df.index[idx], 'StationCode'] = stations[0]['StationCode']
        else:
            st_coords = np.radians(np.array([[s['Latitude'], s['Longitude']] for s in stations]))
            tree = BallTree(st_coords, metric='haversine')
            dist, nearest_idx = tree.query(df_rad[idx], k=1)
            nearest_idx = nearest_idx.flatten()
            df.loc[df.index[idx], 'StationCode'] = [stations[i]['StationCode'] for i in nearest_idx]
    df = df.drop(columns=['StationList'],axis='columns')
    return df
```

```python
new_df = df.copy()
new_df = make_station_info_nearest(new_df, station_df)
display(new_df)
```

```python
# 1. Make date-only columns (timezone removed)
wt_df = new_df.copy()
wt_df['Date'] = wt_df['timestamp'].dt.tz_convert(None).dt.date
water_long['Date'] = pd.to_datetime(water_long['Date'], dayfirst=True).dt.date
# display(wt_df)
# 2. Merge on date and station
wt_df = wt_df.merge(
    water_long,
    how='left',
    left_on=['Date', 'StationCode'],
    right_on=['Date', 'StationCode']
)

display(wt_df)
print(wt_df.isna().sum()) 


```
